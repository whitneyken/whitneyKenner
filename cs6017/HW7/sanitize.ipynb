```python
import pandas as pd

df = pd.read_csv("NP.csv")


#filter only National parks & get unique years

np_df = df.loc[(df["Unit Type"] == "National Park")]
np_df = np_df[np_df['State'] != 'HI']
np_df = np_df[np_df['State'] != 'AK']
df_names = np_df['Unit Name'].unique()
year_df = np_df['YearRaw'].unique()

#pseudo code for restructuring

new_df = pd.DataFrame(index=df_names, columns=year_df).fillna("0")
#for every name in names:
    #for every year in years:
        #make new column in names df
            #add value (0 or value) for each park name and year


# Iterate through each row in np_df and update the corresponding cell in new_df
for i, row in np_df.iterrows():
    park = row['Unit Name']
    year = row['YearRaw']
    string = row[" Visitors "].strip()
    visitors = string.replace(",", "")
    new_df.at[park, year] = visitors


new_df = new_df.reset_index()

new_df.to_json('output.json', orient='records', lines=True)
#new_df.to_csv("output.csv")
```
